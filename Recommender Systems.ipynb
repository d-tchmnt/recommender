```python
#loads the dataset into a dataframe
def load_dataset(str:filepath):
    return pd.read_json(filepath, lines=True)
```

## Loading Data

```python
digital_music_filepath = 'datasets\Digital_Music_5.json'
digital_metadata_filepath = 'datasets\meta_Digital_Music.json'
digital_music = load_dataset(digital_music_filepath)
digital_metadata = load_dataset(digital_metadata_filepath)
```

## Merging

We merge on column 'asin', which is the ID of the product

```python
merged_df = pd.merge(digital_music, digital_metadata, how='inner', on='asin')
merged_df.head(5)
```
